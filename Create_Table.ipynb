# Data Type Table Creation

```python
Data_Types=["numeric", "textual","alphanumeric"]
pickle_in_data_type = open("performance_final_dictionary/new/data_type_score_l.dictionary","rb")
dataType_dict = pickle.load(pickle_in_data_type)
table_data_type = PrettyTable(["Models"]+["Datasets"] + Data_Types)
table_data_type.title = "Performance Matrix For Data Type"
row=['', '', '  P     R     F   cfe ', '  P     R     F   cfe ','  P     R     F   cfe ']
table_data_type.add_row(row)
for model in dataType_dict:
    k=model
    flag=True
    for dataset in dataType_dict[model]:
        if dataset=="wiki":
            continue
        if k and flag:
            row = [model]
            flag=False
        else:
            row=['']
        row.append(dataset)
        for data_type in dataType_dict[model][dataset]:
            score= dataType_dict[model][dataset][data_type]
            #print(data_type)
            if score:
                score=score[0]
                p,r,f,cfe=score
                if p=="Invalid":
                    p,r,f,cfe="null","null","null","null"
            else:
                p,r,f,cfe="","","",""
            row.append("{:}  {:}  {:}  {:}".format(p, r, f,cfe))
        table_data_type.add_row(row)
    row=['------', '------', '------', '------', '------']
    table_data_type.add_row(row)
print(table_data_type)
```

```python
Data_Types=["general"]
pickle_in_data_type = open("performance_final_dictionary/new/general.dictionary","rb")
dataType_dict = pickle.load(pickle_in_data_type)
table_data_type = PrettyTable(["Models"]+["Datasets"] + Data_Types)
table_data_type.title = "Performance Matrix For Data Type"
row=['', '', '  P     R     F   cfe ']
table_data_type.add_row(row)
for model in dataType_dict:
    k=model
    flag=True
    for dataset in dataType_dict[model]:
        if k and flag:
            row = [model]
            flag=False
        else:
            row=['']
        row.append(dataset)
        for data_type in dataType_dict[model][dataset]:
            score= dataType_dict[model][dataset][data_type]
            #print(data_type)
            if data_type=="general":
                if score:
                    score=score[0]
                    p,r,f,cfe=score
                    if p=="Invalid":
                        p,r,f,cfe="null","null","null","null"
                else:
                    p,r,f,cfe="","","",""
                row.append("{:}  {:}  {:}  {:}".format(p, r, f,cfe))
                table_data_type.add_row(row)
    row=['------', '------', '------']
    table_data_type.add_row(row)
    
data = table_data_type.get_string()
with open('dataetype_table.txt', 'w') as f:
    f.write(data)  
print(table_data_type)
```
